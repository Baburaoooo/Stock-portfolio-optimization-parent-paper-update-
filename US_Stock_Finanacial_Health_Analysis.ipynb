```python
import pandas as pd

# Specify the file path
file_path = r'C:\Users\Code file\Stock Health Analysis\Stock_Health_Analysis.csv'

# Read the csv file into a Pandas DataFrame
data = pd.read_csv(file_path, encoding='latin1')

# Display the DataFrame to view the data
data
```

```python
# Remove '%' and 'B' symbols from the specified columns
cols_to_numeric = [
    'Market Cap',
    'EPS Growth 3Y',
    'Rev. Growth 5Y',
    'ROE (5Y)',
    'ROIC (5Y)',
    'GP Growth 5Y',
    'NetInc Growth 5Y',
    'Profit Margin',
    'Debt / Equity',
    'Div. Yield'
]

# Function to remove characters and retain original strings
def clean_specific_characters(column):
    if column.name in cols_to_numeric:
        return column.astype(str).str.replace('%', '').str.replace('B', '')
    return column

# Apply the function to specified columns while keeping other columns as is
data[cols_to_numeric] = data[cols_to_numeric].apply(clean_specific_characters)
```

```python
# Display the DataFrame to view the data
data
```

```python
def pe_ratio(pe_ratio):
    if pe_ratio > 70:
        return 2
    elif 40 <= pe_ratio < 70:
        return 3
    elif 25 <= pe_ratio < 40:
        return 4
    elif 10 <= pe_ratio < 25:
        return 5
    else:
        return 1
```

```python
def eps_growth_rate(eps_growth):
    if eps_growth > 50:
        return 5
    elif 30 <= eps_growth <= 50:
        return 4
    elif 18 <= eps_growth < 30:
        return 3
    elif 10 <= eps_growth < 18:
        return 2
    else:
        return 1
```

```python
def revenue_growth_rate(revenue_growth):
    if revenue_growth > 25:
        return 5
    elif 18 <= revenue_growth <= 25:
        return 4
    elif 13 <= revenue_growth < 18:
        return 3
    elif 8 <= revenue_growth < 13:
        return 2
    else:
        return 1
```

```python
def roe(roe):
    if roe > 30:
        return 2
    elif 22 <= roe <= 30:
        return 4
    elif 14 <= roe < 22:
        return 5
    elif 8 <= roe < 14:
        return 3
    else:
        return 1
```

```python
def roic(roic):
    if roic > 55:
        return 2
    elif 35 <= roic <= 55:
        return 5
    elif 21 <= roic < 35:
        return 4
    elif 6 <= roic < 21:
        return 3
    else:
        return 1
```

```python
def gross_profit_growth_rate(gross_profit_growth):
    if gross_profit_growth > 50:
        return 4
    elif 25 <= gross_profit_growth <= 50:
        return 5
    elif 14 <= gross_profit_growth < 25:
        return 3
    elif 8 <= gross_profit_growth < 14:
        return 2
    else:
        return 1
```

```python
def net_profit_growth_rate(net_profit_growth):
    if net_profit_growth > 50:
        return 4
    elif 35 <= net_profit_growth <= 50:
        return 5
    elif 18 <= net_profit_growth < 35:
        return 3
    elif 12 <= net_profit_growth < 18:
        return 2
    else:
        return 1
```

```python
def net_profit_margin(net_profit_margin):
    if net_profit_margin > 30:
        return 5
    elif 22 <= net_profit_margin <= 30:
        return 4
    elif 13 <= net_profit_margin < 22:
        return 3
    elif 8 <= net_profit_margin < 13:
        return 2
    else:
        return 1
```

```python
def debt_to_equity(debt_to_equity_ratio):
    if debt_to_equity_ratio > 5:
        return 1
    elif 2.5 <= debt_to_equity_ratio <= 5:
        return 2
    elif 1.5 <= debt_to_equity_ratio < 2.5:
        return 3
    elif 0.5 <= debt_to_equity_ratio < 1.5:
        return 5
    else:
        return 4
```

```python
def dividend_yield(dividend_yield):
    if dividend_yield > 5:
        return 1
    elif 3.5 <= dividend_yield <= 5:
        return 2
    elif 2.5 <= dividend_yield < 3.5:
        return 3
    elif 0.7 <= dividend_yield < 2.5:
        return 5
    else:
        return 4
```

```python
# List of columns to convert to numeric type
columns_to_convert = [
    'Market Cap',
    'PE Ratio',
    'EPS Growth 3Y',
    'Rev. Growth 5Y',
    'ROE (5Y)',
    'ROIC (5Y)',
    'GP Growth 5Y',
    'NetInc Growth 5Y',
    'Profit Margin',
    'Debt / Equity',
    'Div. Yield'
]

# Convert specified columns to numeric type
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
from IPython.display import display

# Calculate fitness value for each attribute and create new columns to store the results
data['Fitness_PE_Ratio'] = data['PE Ratio'].apply(pe_ratio)
data['Fitness_EPS_Growth'] = data['EPS Growth 3Y'].apply(eps_growth_rate)
data['Fitness_Rev_Growth'] = data['Rev. Growth 5Y'].apply(revenue_growth_rate)
data['Fitness_ROE_5Y'] = data['ROE (5Y)'].apply(roe)
data['Fitness_ROIC_5Y'] = data['ROIC (5Y)'].apply(roic)
data['Fitness_GP_Growth_5Y'] = data['GP Growth 5Y'].apply(gross_profit_growth_rate)
data['Fitness_NetInc_Growth_5Y'] = data['NetInc Growth 5Y'].apply(net_profit_growth_rate)
data['Fitness_Profit_Margin'] = data['Profit Margin'].apply(net_profit_margin)
data['Fitness_Debt_To_Equity'] = data['Debt / Equity'].apply(debt_to_equity)
data['Fitness_Dividend_Yield'] = data['Div. Yield'].apply(dividend_yield)

# Display the DataFrame with the calculated fitness values for all attributes
display(data)
```

```python
# Save the updated DataFrame with fitness values to a new csv file
output_file_path = r'C:\Users\Code file\Stock Health Analysis\Stock_fitness_values.csv'
data.to_csv(output_file_path, index=True)
```

```python
# Assign weights to each attribute
weights = {
    'weight_pe_ratio': 0.12,
    'weight_eps_growth': 0.12,
    'weight_rev_growth': 0.18,
    'weight_roe_5y': 0.15,
    'weight_roic_5y': 0.05,
    'weight_gp_growth_5y': 0.05,
    'weight_netinc_growth_5y': 0.1,
    'weight_profit_margin': 0.1,
    'weight_debt_to_equity': 0.08,
    'weight_dividend_yield': 0.05
}

# Calculate the combined fitness score for all attributes using weighted sum
data['Combined_Fitness_Score'] = (
    data['Fitness_PE_Ratio'] * weights['weight_pe_ratio'] +
    data['Fitness_EPS_Growth'] * weights['weight_eps_growth'] +
    data['Fitness_Rev_Growth'] * weights['weight_rev_growth'] +
    data['Fitness_ROE_5Y'] * weights['weight_roe_5y'] +
    data['Fitness_ROIC_5Y'] * weights['weight_roic_5y'] +
    data['Fitness_GP_Growth_5Y'] * weights['weight_gp_growth_5y'] +
    data['Fitness_NetInc_Growth_5Y'] * weights['weight_netinc_growth_5y'] +
    data['Fitness_Profit_Margin'] * weights['weight_profit_margin'] +
    data['Fitness_Debt_To_Equity'] * weights['weight_debt_to_equity'] +
    data['Fitness_Dividend_Yield'] * weights['weight_dividend_yield']
)

display(data)
```

```python
# Save the updated DataFrame with fitness values to a new csv file
output_file_path = r'C:\Users\Code file\Stock Health Analysis\Stock_fitness_values.csv'
data.to_csv(output_file_path, index=True)
```
